```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Stage 1": [10.633333, 18.000000, 12.666667, 12.516667, 6.333333, 10.500000, 8.166667, 18.916667, 13.250000],
    "Stage 2": [6.416667, 9.000000, 18.416667, 17.100000, 9.166667, 18.916667, 13.166667, 14.350000, 4.450000]
}

df = pd.DataFrame(data)

plt.figure(figsize=(12, 6))

bar_width = 0.6

bars1 = plt.bar(df.index + 1, df["Stage 1"], label='Stage 1', color='#3b4cc0', width=bar_width)
bars2 = plt.bar(df.index + 1, df["Stage 2"], bottom=df["Stage 1"], label='Stage 2', color='#8db0fe', width=bar_width)

plt.xlabel('Participant')
plt.ylabel('Duration (minutes)')
plt.title('Duration of Stages 1 and 2 per Participant')
plt.rcParams['font.family'] = 'Times New Roman'

plt.xticks(df.index + 1)
plt.legend()

for i, (bar1, bar2) in enumerate(zip(bars1, bars2)):
    plt.text(bar1.get_x() + bar1.get_width() / 2, bar1.get_height() / 2,
             f'{df["Stage 1"][i]:.2f}', ha='center', va='center', color='black', fontsize=10)
    plt.text(bar2.get_x() + bar2.get_width() / 2, bar1.get_height() + bar2.get_height() / 2,
             f'{df["Stage 2"][i]:.2f}', ha='center', va='center', color='black', fontsize=10)
    total_duration = df["Stage 1"][i] + df["Stage 2"][i]
    plt.text(bar2.get_x() + bar2.get_width() / 2, bar1.get_height() + bar2.get_height() + 0.5,
             f'{total_duration:.2f}', ha='center', va='bottom', color='black', fontsize=10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


category_names = ['Strongly disagree', 'Disagree',
                  'Neutral', 'Agree', 'Strongly agree']
results = {
    'had better results': [0,1,3,5,0],
    'was faster': [0,0,0,3,6],
    'was easier': [0,0,0,2,7]
    
}


def survey(results, category_names):

    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    middle_index = data.shape[1]//2
    offsets = data[:, range(middle_index)].sum(axis=1) + data[:, middle_index]/2
    
    category_colors = plt.get_cmap('coolwarm_r')(
        np.linspace(0, 1, data.shape[1]))
    
    fig, ax = plt.subplots(figsize=(10, 2))
    
    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths - offsets
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)
        for rect, width in zip(rects, widths):
            if width > 0:
                x = rect.get_x() + rect.get_width() / 2
                y = rect.get_y() + rect.get_height() / 2
                print(color)
                if colname == "Neutral":
                    ax.text(x, y, int(width), ha='center', va='center', fontsize=9, color='dimgrey')
                else:
                    ax.text(x, y, int(width), ha='center', va='center', fontsize=9, color='white')

    
    ax.axvline(0, linestyle='--', color='black', alpha=.4,zorder=0)
    
    ax.set_xlim(-9, 9)
    ax.set_xticks(np.arange(-9, 10, 1))
    ax.xaxis.set_major_formatter(lambda x, pos: str(abs(int(x))))
    
    ax.invert_yaxis()
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)

    ax.set_title("Stage 2: Compared to doing it without this tool, extracting and validating knowledge...", fontsize=18, pad=40)

    ax.legend(ncol=len(category_names), bbox_to_anchor=(0.5, 1),
          loc='lower center', fontsize='large')
    
    fig.set_facecolor('#FFFFFF')

    return fig, ax

plt.rcParams['font.family'] = 'Times New Roman'
fig, ax = survey(results, category_names)
fig.savefig("stage2.svg", format="svg", bbox_inches='tight')
plt.show()
```
